```python
import threading
import time
import random
from IPython.display import clear_output
```

```python
data = {
    "box1": 0,
    "box2": 0,
    "box3": 0,
    "box4": 0,
    "box5": 0,
    "box6": 0,

}
```

```python
def update_box(name , low , high , refresh_time):
  while True:
    data[name] = random.randint(low , high)
    time.sleep(refresh_time)
```

```python
threads = []

threads.append(threading.Thread(target = update_box, args = ("box1"  , 10 ,20 ,10)))
threads.append(threading.Thread(target = update_box, args = ("box2"  , -10 ,10 ,20)))
threads.append(threading.Thread(target = update_box, args = ("box3"  , -100 ,0 ,8)))
threads.append(threading.Thread(target = update_box, args = ("box4"  , 0 ,20 ,12)))
threads.append(threading.Thread(target = update_box, args = ("box5"  , -40 ,40 ,16)))
threads.append(threading.Thread(target = update_box, args = ("box6"  , 100 ,200 ,14)))
```

```python
for t in threads:
  t.daemon = True
  t.start()
```

```python
start_time = time.time()
while time.time() - start_time < 30:
  clear_output(wait = True)
  print("===== DASHBOARD =====")
  print(f"[10,20]   -> {data['box1']}   (refresh 10s)")
  print(f"[-10,10]  -> {data['box2']}   (refresh 20s)")
  print(f"[-100,0]  -> {data['box3']}   (refresh 8s)")
  print(f"[0,20]    -> {data['box4']}   (refresh 12s)")
  print(f"[-40,40]  -> {data['box5']}   (refresh 16s)")
  print(f"[100,200] -> {data['box6']}   (refresh 14s)")

  time.sleep(1)
```
