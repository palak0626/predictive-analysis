```python
import threading
```

```python
def create_file(filename , content):
  with open(filename , 'w') as f:
    f.write(content)
  print(f"{filename} created")
```

```python
threads = []

threads.append(threading.Thread(target = create_file , args = ("file1.txt" , "hello from file 1")))
threads.append(threading.Thread(target = create_file , args = ("file2.txt" , "hello form file 2")))
threads.append(threading.Thread(target = create_file , args = ("file3.txt" , "hello from file 3")))
threads.append(threading.Thread(target = create_file , args = ("file3.txt" , "hello from file 4")))
threads.append(threading.Thread(target = create_file , args = ("file3.txt" , "hello from file 5")))
```

```python
for t in threads:
  t.start()
```

```python
for t in threads:
  t.join()
```
